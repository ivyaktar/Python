```python
#Task1

class Student:
  def __init__(self,name,ID,dep='CSE'):
    self.name = name
    self.ID = ID
    self.dep = dep
    print(f'Name: {self.name}')
    print(f'ID: {self.ID}')
    print(f'Department: {self.dep}')

  def dailyEffort(self,de):
    self.de = de
    print(f'Daily Effort: {self.de} hour(s)')
  def printDetails(self):
    if self.de <= 2:
      print(f'Suggestion: Should give more effort!')
    elif self.de <= 4:
      print(f'Suggestion: Keep up the good work!')
    else:
      print(f'Suggestion: Excellent! Now motivate others.')

harry = Student('Harry Potter', 123)
harry.dailyEffort(3)
harry.printDetails()
print('========================')
john = Student("John Wick", 456, "BBA")
john.dailyEffort(2)
john.printDetails()
print('========================')
naruto = Student("Naruto Uzumaki", 777, "Ninja")
naruto.dailyEffort(6)
naruto.printDetails()
```

```python
#Task2

class Farmer:
  def __init__(self,name=None):
    if name == None:
      print('Welcome to your farm!')
    elif type(name) == str:
      print(f'Welcome to your farm, {name}!')
    elif type(name) == int:
      print(f'Welcome to your farm. Your famr ID is {name}!')

    self.crops = []
    self.fishes = []

  def addCrops(self,*crops):
    self.crops += list(crops)
    if self.crops != []:
      print(f'{len(crops)} crop(s) added.')
    else:
      print(f'No crop(s) added.')

  def addFishes(self,*fishes):
    self.fishes += list(fishes)
    if self.fishes != []:
      print(f'{len(fishes)} fish(s) added.')
    else:
      print(f'No fish(s) added.')

  def showGoods(self):
    if self.crops != []:
      print(f'You have {len(self.crops)} crop(s):')
      os = ','.join(self.crops)
      print(os)
    else:
      print("You don't have any crop(s)")

    if self.fishes != []:
      print(f'You have {len(self.fishes)} fish(s):')
      os = ','.join(self.fishes)
      print(os)
    else:
      print(f"You don't have any fish(s)")

#DriverCode
f1 = Farmer()
print("-------------------")
f1.addCrops('Rice', "Jute", "Cinnamon")
print("-------------------")
f1.addFishes()
print("-------------------")
f1.addCrops('Mustard')
print("-------------------")
f1.showGoods()
print("-------------------")
f2 = Farmer("Korim Mia")
print("-------------------")
f2.addFishes('Pangash', 'Magur')
print("-------------------")
f2.addCrops("Wheat", "Potato")
print("-------------------")
f2.addFishes("Koi", "Tuna", "Sardine")
print("-------------------")
f2.showGoods()
print("-------------------")
f3 = Farmer(2865127000)
print("-------------------")
f3.addCrops()
print("-------------------")
f3.addFishes("Katla")
print("-------------------")
f3.showGoods()
print("-------------------")
```

```python
#Task3

class TaxiLagbe:
  def __init__(self,tn,area):
    self.tn = tn
    self.area = area
    self.pas = []
    self.tf = 0
    self.namel = []
  def addPassenger(self,*pas):

    if len(self.pas) >= 4:
      print('Taxi Full! No more passengers can be added.')
    else:
      self.pas += list(pas)
      for i in pas:
        templ = i.split('_')
        self.namel += [templ[0]]
        self.tf += int(templ[1])
        print(f'Dear {templ[0]}! Welcome to TaxiLagbe.')

  def printDetails(self):
    print(f'Trip info for Taxi number: {self.tn}\nThis taxi can only cover the {self.area} area.')
    print(f'Total passengers: {len(self.pas)}')
    os = ','.join(self.namel)
    print(f'Passenger list: {os}')
    print(f'Total collected fare: {self.tf} Taka')

#DriverCode
taxi1 = TaxiLagbe('1010-01', 'Dhaka')
print('-------------------------------')
taxi1.addPassenger('Walker_100', 'Wood_200','Matt_100')
taxi1.addPassenger('Wilson_105')
print('-------------------------------')
taxi1.printDetails()
print('-------------------------------')
taxi1.addPassenger('Karen_200')
print('-------------------------------')
taxi1.printDetails()
print('-------------------------------')
taxi2 = TaxiLagbe('1010-02', 'Khulna')
taxi2.addPassenger('Ronald_115', 'Parker_215')
print('-------------------------------')
taxi2.printDetails()
```

```python
#Task4
import math

class Sphere:
  def __init__(self,sID,r=1,color='white'):
    self.ID = sID
    self.r = r
    self.c = color
    self.v = (4/3)*math.pi*(self.r**3)
  def printDetails(self):
    print(f'Sphere ID: {self.ID}')
    print(f'Color: {self.c}')
    print(f'Volume: {self.v}')
  def merge_sphere(self,*spheres):
    print('Spheres are being merged.')
    for i in spheres:
      if i.c != self.c:
        self.c = 'Mixed Color'
      self.v += i.v

#DriverCode
sphere1 = Sphere("Sphere 1")
print("1***************")
sphere1.printDetails()
print("2***************")
sphere2 = Sphere("Sphere 2", 3)
print("3***************")
sphere2.printDetails()
print("4***************")
sphere3 = Sphere("Sphere 3", 2)
print("5***************")
sphere3.printDetails()
print("6***************")
sphere3.merge_sphere(sphere1,sphere2)
print("7***************")
sphere3.printDetails()
print("8***************")
sphere4 = Sphere("Sphere 4", 5, "Purple")
print("9***************")
sphere4.merge_sphere(sphere3)
print("10***************")
sphere4.printDetails()
```
