```python
#TASK 7 (For your convenience, see the ascii codes/values on the website mentioned in Task-7 question in Assignment 4)

# Ask the user to enter a string
a = input()

# Assign a variable to store the ASCII value of characters
b = 0

# Go through each character in the input string
for i in a:

    # Get the ASCII value of the current character
    ascii = ord(i)

    # Check if the ASCII value is greater than or equal to 122 ('z=122')
    if ascii >= 122:

        # If the ASCII value is 'z' or greater, set it to 97 ('a=97')
        ascii = 97

        # Print the corresponding character using the updated ASCII value, without a newline
        print(chr(ascii), end="")
        # The "chr()" function is used to convert an ASCII value to its corresponding character.

    else:

        # If the ASCII value is less than 'z', add 1 to it
        b = ascii + 1

        # Print the corresponding character using the updated ASCII value, without a newline
        print(chr(b), end="")
```

```python
#TASK 10

# Ask the user to enter a string consisting of two parts separated by a comma and space
a = input()

# Find the index of the comma and space in the input string
comma_index = a.index(", ")

# Extract and assign the first part of the string before the comma and space
string1 = a[:comma_index]

# Extract and assign the second part of the string after the comma and space
string2 = a[comma_index + 2:]

# Create an empty string variable to store the output
output = ""

# Compare the lengths of string1 and string2 to determine the shorter length
if len(string1) < len(string2):

    # If string1 is shorter, assign its length to shorter_length and assign string2 to longer_string
    shorter_length = len(string1)
    longer_string = string2

else:

    # If string2 is shorter or they have equal lengths, assign string2's length to shorter_length and string1 to longer_string
    shorter_length = len(string2)
    longer_string = string1

# Go through each character in the  "shorter_length" variable
for i in range(shorter_length):

    # Add the characters at the corresponding positions from string1 and string2 and add them to the output
    output += string1[i] + string2[i]

# Add the remaining characters from the longer string to the output
output += longer_string[shorter_length:]

# Print the final output
print(output)




```
