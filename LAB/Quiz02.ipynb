```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class QuizStack:
    def __init__(self):
        self.top = None  # Represents the top of the stack--head here
        self.size = 0    # Tracks the size of the stack

    def isEmpty(self):
        return self.size == 0

    def push(self, data): # Prepend--head changes
        new_node = Node(data)
        new_node.next = self.top  # New node points to the current top
        self.top = new_node       # Update the top to the new node
        self.size += 1            # Increment size

    def pop(self): # remove top head--index 0
        if self.isEmpty():
            return None
        popped_data = self.top.data  # Retrieve the data of the top node
        self.top = self.top.next     # Update the top to the next node
        self.size -= 1               # Decrement size
        return popped_data

    def peek(self):
        if self.isEmpty():
            return None
        return self.top.data
```

```python
def find_second_largest(stack):
    if stack.isEmpty():
        return None  # Edge case (though the problem assumes at least 2 elements)

    # Initialize the two largest elements to None initially
    largest = None
    second_largest = None

    # Traverse through the stack
    while not stack.isEmpty():
        value = stack.pop()

        # Update largest and second largest
        if largest is None or value > largest:
            second_largest = largest
            largest = value
        elif second_largest is None or value > second_largest:
            second_largest = value

    return second_largest


quiz_stack = QuizStack()
for num in [7, 10, 5, 5, 40, -3, 30, 10]:
    quiz_stack.push(num)

print(find_second_largest(quiz_stack))  # Output: 30


quiz_stack = QuizStack()
for num in [3, 5, 2, 1, -5, 1]:
    quiz_stack.push(num)

print(find_second_largest(quiz_stack))  # Output: 3
```

```python
def find_second_smallest(stack):
    if stack.isEmpty():
        return None  # Edge case (though the problem assumes at least 2 elements)

    # Initialize the two smallest elements to None initially
    smallest = None
    second_smallest = None

    # Traverse through the stack
    while not stack.isEmpty():
        value = stack.pop()

        # Update smallest and second smallest
        if smallest is None or value < smallest:
            second_smallest = smallest
            smallest = value
        elif second_smallest is None or value < second_smallest:
            second_smallest = value

    return second_smallest


# Example 1:
quiz_stack = QuizStack()
for num in [3, 5, 2, 1, -5, 10]:
    quiz_stack.push(num)
print(find_second_smallest(quiz_stack))  # Output: 1

# Example 2:
quiz_stack = QuizStack()
for num in [7, 10, 5, 15, 40, -3, 30, 10]:
    quiz_stack.push(num)
print(find_second_smallest(quiz_stack))  # Output: 5
```
