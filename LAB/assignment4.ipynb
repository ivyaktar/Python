```python
#task_3

def check_binary_number(input_string):
    for char in input_string:
        if char != '0' and char != '1':
            return False
    return True


input_string = input("Enter a string: ")


is_binary = check_binary_number(input_string)


if is_binary:
    print("Binary Number")
else:
    print("Not a Binary Number")
```

```python
#task_7


string1 = input("Enter a string: ")


string2 = ""


for n in string1:

    if 'a' <= n <= 'z':

        code1 = ord(n)

        code2 = code1 + 1

        if code2 > ord('z'):

            code2 = ord('a')

        n1 = chr(code2)

        string2 += n1
    else:

        string2 += n

print(string2)

```

```python
#task_8


input_string = input("Enter a string: ")


output_string = ""


for index, char in enumerate(input_string):

    if index % 2 != 0:

        output_string += char


uppercase_string = ""
for char in output_string:

    if 'a' <= char <= 'z':

        ascii_code = ord(char)

        uppercase_code = ascii_code - 32

        uppercase_char = chr(uppercase_code)

        uppercase_string += uppercase_char
    else:

        uppercase_string += char


print(uppercase_string)
```

```python
# Task 10

input_string = input("Please enter two strings separated by a comma and a space: ")


string1, string2 = input_string.split(", ")


mixed_string = ""
length1 = len(string1)
length2 = len(string2)


max_length = max(length1, length2)


for i in range(max_length):

    if i < length1:
        mixed_string += string1[i]

    if i < length2:
        mixed_string += string2[i]


if length1 > length2:
    mixed_string += string1[length2:]
else:
    mixed_string += string2[length1:]


print(mixed_string)
```
