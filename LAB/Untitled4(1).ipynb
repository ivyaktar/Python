    (#1) Assign the arguments into appropriate attributes: data, talk_time, messages, cashback, validity and price via a parameterized constructor. All the attributes should be of integer data type. Note that data is stored in Megabytes (1 GB = 1024 MB) and the cashback amount is calculated from a percentage value.
     
    (#2,3,4) Implement driver code to display all the information of a package. Check if any particular attribute does not exist (is equal to 0), do not print that attribute. Attributes validity and price are always printed.




````python
# Driver Code
# Subtask 1: Write the CellPackage Class

#(price, data, talk_time, sms, cashback_percent,validity)

class CellPackage:
  def __init__(self, price, data, talk_time, sms, cashback_percent, validity):
    self.price=price
    data=data.split(' ')
    self.data=int(data[0])*1024          #1 GB = 1024 MB
    self.talk_time=talk_time
    self.sms=sms
    c_p=int(cashback_percent[0:len(cashback_percent)-1])
    self.cashback=int((c_p*price)/100)
    self.validity=validity

  def show(self):
    if self.data!=0:
      print(f'Data = {self.data}')

    if self.talk_time!=0:
      print(f'Talktime = {self.talk_time}')

    if self.sms!=0:
      print(f'SMS/MMS = {self.sms}')

    print(f'Validity = {self.validity}')
    print(f'--> Price = {self.price}')

    if self.cashback!=0:
      print(f'Buy now to get {self.cashback} tk cashback.')

pkg = CellPackage(150, '6 GB', 99, 20, '7%', 7)
print('============= Package 1 =============')
# Subtask 2: Check each attribute and print
pkg.show()

pkg2 = CellPackage(700, '35 GB', 700, 0, '10%', 30)
print('============= Package 2 =============')
# Subtask 3: Check each attribute and print
pkg2.show()

pkg4 = CellPackage(120, '0 GB', 190, 0, '0%', 10)
print('============= Package 3 =============')
# Subtask 4: Check each attribute and print
pkg4.show()

````

Your task is to write the driver code that:
Creates 2 Vector3D objects.
    The first one is given as V1=<2, -3, 1> and the second one is given as V2=<-1, 4, 0>.
    
    Prints their components and magnitude as shown in the output.
    
    The magnitude of a vector <a, b, c> is calculated as |<a, b, c> | = a2 + b2 + c2.
    
    Finds the dot product of the 2 Vectors. Dot product of 2 vectors <a1, a2, a3> and <b1, b2, b3> is calculated as <a1, a2, a3>  <b1, b2, b3> = a1 b1 + a2 b2 + a3 b3.
    
    Finds the Cross product of the 2 Vectors. The cross product of 2 vectors <a1, a2, a3> and <b1, b2, b3> creates a new Vector3D Object which is calculated as <a1, a2, a3>  <b1, b2, b3> = <a2b3 - a3b2, a3b1 - a1b3, a1b2 - a2b1>.

    Generates the output as given below.

    Your program should run for any two 3D vectors.




```
Vector <2, -3, 1> has been created.
Vector <-1, 4, 0> has been created.
Magnitude of the first vector = 3.7416573867739413
Magnitude of the second vector = 4.123105625617661
Dot product of the two vectors = -14
Vector <-4, -1, 5> has been created.
Cross product of the two vectors = <-4, -1, 5>

```



````python
import math
class Vector3D:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z
        print(f'Vector <{self.x}, {self.y}, {self.z}> created.')

v1= Vector3D(2, -3, 1)
v2= Vector3D(-1, 4, 0)

print(f'Magnitude of the first vector = {math.sqrt(v1.x**2+v1.y**2+v1.z**2)}')
print(f'Magnitude of the second vector = {math.sqrt(v2.x**2+v2.y**2+v2.z**2)}')

print(f'Dot product of the two vectors = {v1.x*v2.x+v1.y*v2.y+v1.z*v2.z}')

#a2b3 - a3b2, a3b1 - a1b3, a1b2 - a2b1
v3= Vector3D(v1.y*v2.z-v1.z*v2.y, v1.z*v2.x-v1.x*v2.z, v1.x*v2.y-v1.y*v2.x)

print(f'Cross Product of the two vectors = <{v3.x}, {v3.y}, {v3.z}>')
````

Task-4

Design the following abcTech class so that it generates the following output:
Hints:
   If the working hour of the employee is more than 144 hours in the month, then he/she will be paid Tk. 800 for each extra hour worked along with the base salary.

   If the working hour of the employee is less than or equal to 144 hours, then the salary will be the same as the base salary.  




```
-------------------------
Welcome to abcTech, Tamim Hasan!
-------------------------
Name:  Tamim Hasan
Designation:  Software Engineer
Department: Android Development
Programming Skills:  Java, Python, Dart, C++
Frameworks:  Express.js, React
-------------------------
Your salary for this month is Tk. 54600
-------------------------
-------------------------
Welcome to abcTech, Jahin Khandoker!
-------------------------
Name:  Jahin Khandoker
Designation:  Senior Developer
Department:  App Development
Programming Skills:  'Java', 'Dart', 'Swift'
Frameworks:  'Flutter', 'React Native', 'Xamarin'
-------------------------
Your salary for this month is Tk. 103000
-------------------------
```



````python
class abcTech:
  def __init__(self, a, b, c):
    self.name=a
    self.designation=b
    self.department=c
    self.programmingskill=''
    self.framework=''
    print(f'Welcome to abcTech, {a}!')

  def addProgrammingSkills(self, li):
    for i in li:
      if len(self.programmingskill)==0:
        self.programmingskill+=i
      else:
        self.programmingskill+=', '+i

  def addFrameworks(self, li):
    for i in li:
      if len(self.framework)==0:
        self.framework+=i
      else:
        self.framework+=', '+i

  def printInfo(self):
    print(f'Name: {self.name}')
    print(f'Designation: {self.designation}')
    print(f'Department: {self.department}')
    print(f'Programming Skills: {self.programmingskill}')
    print(f'Frameworks: {self.framework}')

  def calculateSalary(self, base, hr):
    if hr>144:
      extra=hr-144
      return f'{base+(extra*800)}'
    else:
      return f'{base}'

print("-------------------------")
b1 =abcTech("Tamim Hasan", "Software Engineer", "Android Development")
print("-------------------------")
b1.addProgrammingSkills(["Java", "Python"])
b1.addProgrammingSkills(["Dart", "C++"])
b1.addFrameworks(["Express.js", "React"])
b1.printInfo()
print("-------------------------")
print(f"Your salary for this month is Tk. {b1.calculateSalary(45000, 156)}")
print("-------------------------")
print("-------------------------")
b2 =abcTech("Jahin Khandoker", "Senior Developer", "App Development")
print("-------------------------")
b2.addProgrammingSkills(["Java", "Dart","Swift"])
b2.addFrameworks(["Flutter", "React Native"])
b2.addFrameworks(["Xamarin"])
b2.printInfo()
print("-------------------------")
print(f"Your salary for this month is Tk. {b2.calculateSalary(103000, 123)}")
print("-------------------------")
````
