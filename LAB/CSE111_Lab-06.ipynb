```python
#Task_1
class Student:
  ID = 0
  def __init__(self, name, department, age, cgpa):
    Student.ID +=1
    self.id = Student.ID
    self.name = name
    self.dept = department
    self.age = age
    self.cg = cgpa

  def showDetails(self):
    print(f'''ID: {self.id}
Name: {self.name}
Age: {self.age}
CGPA: {self.cg}''')

  @classmethod
  def from_String(cls, x):
    y = x.split('-')
    return cls(y[0], y[1], y[2], y[3])

s1 = Student("Samin", "CSE", 21, 3.91)
s1.showDetails()
print("-----------------------")
s2 = Student("Fahim", "ECE", 21, 3.85)
s2.showDetails()
print("-----------------------")
s3 = Student("Tahura", "EEE", 22, 3.01)
s3.showDetails()
print("-----------------------")
s4 = Student.from_String("Sumaiya-BBA-23-3.96")
s4.showDetails()

print("Instance variables are diffrent for objects of a class. But class variables are same for all the objects of a class.")
print("Instance methods works when an object is created, while class methods works on the class itself and can be called on the class without creating an object")
```

```python
#Task_2
class Passenger:
  count = 0

  def __init__(self, n):
    Passenger.count+=1
    self.name = n
    self.fare = 450
    self.weight = 0

  def set_bag_weight(self, x):
    self.weight = x
    if self.weight > 50:
      self.fare += 100
    elif self.weight > 20:
      self.fare += 50

  def printDetail(self):
    print(f'''Name: {self.name}
Bus Fare: {self.fare} taka''')


print("Total Passenger:", Passenger.count)
p1 = Passenger("Jack")
p1.set_bag_weight(90)
p2 = Passenger("Carol")
p2.set_bag_weight(10)
p3 = Passenger("Mike")
p3.set_bag_weight(25)
print("=========================")
p1.printDetail()
print("=========================")
p2.printDetail()
print("=========================")
p3.printDetail()
print("=========================")
print("Total Passenger:", Passenger.count)
```

```python
#Task_4

class NikeBangladesh:
  branch = []
  stock = {'Air Jordan': 0, 'Cortez': 0, 'Zoom Kobe' : 0}
  sold = 0

  def __init__(self, n):
    self.name = n
    self.stock = {'Air Jordan': 0, 'Cortez': 0, 'Zoom Kobe' : 0}
    self.sold = 0
    NikeBangladesh.branch.append(self.name)

  def details(self):
    print(f'''Nike {self.name} outlet:
Products Currently Stocked
{self.stock}
Sold: {self.sold}''')

  def restockProducts(self, x):
    for i, j in x.items():
      self.stock[i] += j
      NikeBangladesh.stock[i] += j

  def productSold(self, x):
    for i, j in x.items():
      self.stock[i] -= j
      NikeBangladesh.stock[i] -= j
      self.sold += j
      NikeBangladesh.sold += j

  @classmethod
  def status(cls):
    print(f'''Nike Bangladesh Status:
Branches Opened: {cls.branch}
Currently Stocked
{cls.stock}
Sold: {cls.sold}''')


print("xxxxxxxxxxxxxx1xxxxxxxxxxxxxxxx")
NikeBangladesh.status()
dhaka = NikeBangladesh("Dhaka Banani")
chittagong = NikeBangladesh("Chittagong GEC")
print("xxxxxxxxxxxxxx2xxxxxxxxxxxxxxxx")
dhaka.details()
print("xxxxxxxxxxxxxx3xxxxxxxxxxxxxxxx")
chittagong.details()
print("xxxxxxxxxxxxxx4xxxxxxxxxxxxxxxx")
dhaka.restockProducts(
{"Air Jordan":1200,"Cortez":200,"Zoom Kobe":200})
chittagong.restockProducts(
{"Air Jordan":1000,"Cortez":250,"Zoom Kobe":100})
print("xxxxxxxxxxxxxx5xxxxxxxxxxxxxxxx")
NikeBangladesh.status()
print("xxxxxxxxxxxxxx6xxxxxxxxxxxxxxxx")
dhaka.productSold({"Air Jordan":760,"Cortez":90})
chittagong.productSold({"Air Jordan":520,"Zoom Kobe":70})
print("xxxxxxxxxxxxxx7xxxxxxxxxxxxxxxx")
NikeBangladesh.status()
```

```python
#Task_6
class SultansDine:
    branch = 0
    branch_list = []
    sell = 0

    def __init__(self, n):
        self.name = n
        self.sell = 0
        SultansDine.branch_list.append(self)
        SultansDine.branch += 1

    def sellQuantity(self, x):
        if x < 10:
            self.sell = x * 300
        elif x < 20:
            self.sell = x * 350
        else:
            self.sell = x * 400

        SultansDine.sell += self.sell

    @classmethod
    def details(cls):

        print(f'''Total Number of branch(s): {SultansDine.branch}
Total Sell: {cls.sell} Taka''')

        if SultansDine.branch != 0:
            for i in SultansDine.branch_list:
                print(f'''Branch Name: {i.name}, Branch Sell: {i.sell} Taka
Branch consists of total sell's: {round(((i.sell / cls.sell)*100),2)}%''')
    def branchInformation(self):
      print(f'''Branch Name: {self.name}
  Branch Sell: {self.sell} Taka''')

SultansDine.details()
print('########################')
dhanmondi = SultansDine('Dhanmondi')
dhanmondi.sellQuantity(25)
dhanmondi.branchInformation()
print('-----------------------------------------')
SultansDine.details()

print('========================')

baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()

print('========================')

gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```
