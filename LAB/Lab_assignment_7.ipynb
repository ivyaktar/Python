```python
#Task2

def fibonacci(range):
  fib1 = 0
  fib2 = 1
  while fib1 <= range:
    fib3 = fib1 + fib2
    print(fib1, end = ' ')
    fib1 = fib2
    fib2 = fib3
fibonacci(10)
```

```python
#Task4

def case_check(s):
  uppercase = 0
  lowercase = 0

  for i in s:
    if 97 <= ord(i) <= 122:
      lowercase += 1
    if 65 <= ord(i) <= 90:
      uppercase += 1

  print("No. of Uppercase character: ", uppercase)
  print("No. of Lowercase character: ", lowercase)

case_check('HaRRy PotteR')
```

```python
#Task5

def calculate_tax(age, salary, job):

  job1 = ''
  for i in job:
    if ord(i) < 97:
      job1 += chr(ord(i)+32)

  if age < 18 or job1 == 'president' or salary < 10000:
    return 0
  elif 10000 < salary < 20000 :
    return (salary*(5/100))
  elif salary > 20000:
    return (salary*(10/100))

calculate_tax(20, 22000, 'Assistant manager')
```

```python
#Task7

def show_palindrome(x):
  pal = ''
  for i in range(1,x+1):
    pal += str(i)
  for j in range(x-1,0,-1):
    pal += str(j)
  return pal
print(show_palindrome(1))
```

```python
#Task13

def calc(o,n1,n2):

  if o == '+':
    c = n1 + n2
  elif o == '-':
    c = n1 - n2
  elif o == '/':
    c = n1 / n2
  elif o == '*':
    c = n1 * n2
  return c

o_i = input()
n1_i = float(input())
n2_i = float(input())
print(calc(o_i,n1_i,n2_i))
```

```python
#Task14

def chr_remove(s,p):

  out = ''
  temp = ''
  for i in range(len(s)):

    if i != 0:
      if i % p != 0:
        out += s[i]
      else:
        temp += s[i]
    else:
      out += s[i]

  out += temp
  return out
s_i = input()
p_i = int(input())
print(chr_remove(s_i,p_i))
```

```python
#Task15

def price(o,l='Dhanmondi'):
  dict1 = {'Rice':105,'Potato':20,'Chicken':250,'Beef':510,'Oil':85}
  price = 0
  for i in o:
    price += dict1[i]

  if l == 'Dhanmondi':
    price += 30
  else:
    price += 70

  return price

print(price(["Rice", "Beef", "Rice"]))
```

```python
#Task18

def division(sv,ev,fd,sd):
  total = 0
  for i in range(sv,ev):
    if i % fd == 0 and i % sd == 0:
      pass
    else:
      if i % fd == 0 or i % sd == 0:
        total += i
  return total

sv = int(input())
ev = int(input())
fd = int(input())
sd = int(input())

print(division(sv,ev,fd,sd))
```
