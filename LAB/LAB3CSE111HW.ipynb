```python
#Task1
# Driver Code
class CellPackage:
  def __init__(self,price,data,talk_time,messages, cashback, validity):
    self.price=price
    self.data=int(data.split(" GB")[0])
    self.talk_time=talk_time
    self.messages=messages
    self.cashback=int(cashback.split("%")[0])
    self.validity=validity

  def show_details(self):
    if self.data!=0:
      print(f"Data = {self.data*1024} MB")
    if self.talk_time!=0:
      print(f"Talktime = {self.talk_time} Minutes")
    if self.messages!=0:
      print(f"SMS/MMS = {self.messages}")
    print(f"Validity = {self.validity} days")
    print(f"--> Price = {self.price}")
    if self.cashback !=0:
      print(f"Buy now to get {int(self.price*(self.cashback/100))} tk cashback.")
pkg = CellPackage(150, '6 GB', 99, 20, '7%', 7)
print('============= Package 1 =============')
pkg.show_details()
pkg2 = CellPackage(700, '35 GB', 700, 0, '10%', 30)
print('============= Package 2 =============')
pkg2.show_details()
pkg4 = CellPackage(120, '0 GB', 190, 0, '0%', 10)
print('============= Package 3 =============')
pkg4.show_details()
```

```python
#Task2
#class code
class box:
  def __init__(self, valList):
    self.height, self.width, self.breadth = valList
    print("Creating a box")
  def boxDescription(self):
    print(f"Height = {self.height}")
    print(f"Width = {self.width}")
    print(f"Breadth = {self.breadth}")
  def volume(self):
    self.volume= self.height*self.width*self.breadth

# drivercode
print("Box 1")
b1 = box([10,10,10])
print("=========================")
b1.boxDescription()
print(b1.volume())
print(f"Volume of the box is {volume} cubic units.")
print("-------------------------")
print("Box 2")
b2 = box((30,10,10))
print("=========================")
b2.boxDescription()
print(b2.volume())
print(f"Volume of the box is {volume} cubic units.")
b2.height = 300
print("Updating Box 2!")
print("Height:", b2.height)
print("Width:", b2.width)
print("Breadth:", b2.breadth)
volume = b2.height * b2.width * b2.breadth
print(f"Volume of the box is {volume} cubic units.")
print("-------------------------")
print("Box 3")
b3 = b2
b3.boxDescription()
print(b3.volume())
print(f"Volume of the box is {volume} cubic units.")
```

```python
#Task4
#class code
class abcTech:
  def __init__(self, name,job,dept):
    self.name = name
    self.job = job
    self.dept = dept
    self.skills=None
    self.frameworks = None
    print(f"Welcome to abcTech, {self.name}! ")
  def addProgrammingSkills (self, skills):
    self.skills = skills
  def addFrameworks (self,frameworks):
    self.frameworks = frameworks
  def printInfo(self):
    print(f"""Name: {self.name}
          Designation: {self.job}
          Department: {self.dept}
          Programming Skills: {", ".join(self.skills)}
          Frameworks: {", ".join(self.frameworks)} """)
  def calculateSalary(self, salary, hours):
    if hours > 144:
        salary+=(800*(hours-144))
    else:
        pass
    return salary

#Driver code
print("-------------------------")
b1 =abcTech("Tamim Hasan", "Software Engineer", "Android Development")
print("-------------------------")
b1.addProgrammingSkills(["Java","Python"])
b1.addProgrammingSkills(["Dart", "C++"])
b1.addFrameworks(["Express.js", "React"])
b1.printInfo()
print("-------------------------")
print(f"Your salary for this month is Tk.{b1.calculateSalary(45000, 156)}")
print("-------------------------")
print("-------------------------")
b2 =abcTech("Jahin Khandoker", "Senior Developer", "App Development")
print("-------------------------")
b2.addProgrammingSkills(["Java", "Dart","Swift"])
b2.addFrameworks(["Flutter", "React Native"])
b2.addFrameworks(["Xamarin"])
b2.printInfo()
print("-------------------------")
print(f"Your salary for this month is Tk.{b2.calculateSalary(103000, 123)}")
print("-------------------------")
```

```python
#Task5
class StudentDatabase:
    def __init__(self, name, id):
        self.name = name
        self.id = id
        self.grades = {}

    def calculateGPA(self, course, semester):
        total_credits = 0
        courses_tup = []
        for i in course:
            j = i.split(": ")
            total_credits += float(j[1])
            courses_tup.append(j[0])
        cgpa = round(total_credits / len(course), 2)
        sem_grades = {tuple(courses_tup): cgpa}
        self.grades[semester] = sem_grades

    def printDetails(self):
        print(f"Name: {self.name}")
        print(f"ID: {self.id}")
        for semester, courses in self.grades.items():
            print(f"Courses taken in {semester}: ")
            for course_names, cgpa in courses.items():
                for course in course_names:
                    print(course)
            print(f"GPA: {cgpa}")
            print('---------------------------------')


s1 = StudentDatabase('Pietro', '10101222')
s1.calculateGPA(['CSE230: 4.0', 'CSE220: 4.0', 'MAT110: 4.0'], 'Summer2020')
s1.calculateGPA(['CSE250: 3.7', 'CSE330: 4.0'], 'Summer2021')
print(f'Grades for {s1.name}\n{s1.grades}')
print('---------------------------------')
s1.printDetails()
s2 = StudentDatabase('Wanda', '10103332')
s2.calculateGPA(['CSE111: 3.7', 'CSE260: 3.7', 'ENG101: 4.0'], 'Summer2022')
print('---------------------------------')
print(f'Grades for {s2.name}\n{s2.grades}')
print('---------------------------------')
s2.printDetails()
```

```python
class Store:
    def __init__(self, initial_balance):
        print("New branch created!")
        self.balance = initial_balance
        self.inventory = {}

    def add_item(self, item_info):
        item_name, stock, buying_price, selling_price = item_info
        if item_name in self.inventory:
            self.inventory[item_name]["stock"] += stock
        else:
            self.inventory[item_name] = {
                "stock": stock,
                "buying_price": buying_price,
                "selling_price": selling_price,
            }
        print(f"Item added: {item_name}")
        self.balance -= stock * buying_price

    def sell_item(self, item_name, quantity):
        if item_name in self.inventory and self.inventory[item_name]["stock"] >= quantity:
            selling_price = self.inventory[item_name]["selling_price"]
            self.inventory[item_name]["stock"] -= quantity
            revenue = quantity * selling_price
            self.balance += revenue
            print(f"Sold {quantity} {item_name}(s) for ${revenue}.")
        else:
            print(f"Sorry! {item_name} is not available at your desired quantity. Currently we have: {self.inventory.get(item_name, {}).get('stock', 0)}")

    def restock_item(self, item_name, quantity):
        if item_name in self.inventory:
            self.inventory[item_name]["stock"] += quantity
            print(f"Restocked item: {item_name}, Current Stock: {self.inventory[item_name]['stock']}")
        else:
            print(f"Item {item_name} not found in inventory.")

    def viewAllItems(self):
        items = ", ".join(self.inventory.keys())
        print(f"All Items: {items}")

    def viewAllItemDetails(self):
        print(self.inventory)


# Driver Code
print("==========================")
branch1 = Store(5000)
print(f"Current Balance: {branch1.balance}")
print(f"Total items: {len(branch1.inventory)}")
branch1.viewAllItems()
branch1.viewAllItemDetails()
print("==========================")
print(f"Current Balance: {branch1.balance}")
branch1.add_item(["ChaCha Noodles", 10, 5, 8])
print(f"Current Balance: {branch1.balance}")
branch1.add_item(["Sparrow Shampoo", 5, 10, 20])
print(f"Current Balance: {branch1.balance}")
print("==========================")
branch1.viewAllItems()
print()
branch1.viewAllItemDetails()
print()
print("==========================")
print(f"Current Balance: {branch1.balance}\n")
branch1.sell_item("ChaCha Noodles", 15)
print(f"Current Balance: {branch1.balance}\n")
branch1.viewAllItemDetails()
print()
branch1.sell_item("ChaCha Noodles", 10)
print()
print(f"Current Balance: {branch1.balance}\n")
branch1.viewAllItemDetails()
print()
print("==========================")
print(f"Current Balance: {branch1.balance}\n")
branch1.restock_item("ChaCha Noodles", 5)
print()
branch1.viewAllItemDetails()
print()
print(f"Current Balance: {branch1.balance}\n")
print("==========================")
```
