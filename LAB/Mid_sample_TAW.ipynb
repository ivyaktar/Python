```python
# Get the number of games played from the user
games = int(input())

# Assign variables to store total points and number of wins.
points = 0
wins = 0

# Loop through each game given by the user to collect scores and calculate points.
for i in range(games):

    # Get the scores for the current game from the user in the format: team_score-opponent_score.
    # We will use '5-2' as an example for this explanation. Suppose the user inputs 5-2, then from string indexing method, the index of '5' is 0, index of 'hyphen('-')' is 1 and index of '2' is 2.
    scores = input()

    # Now, from previous explanation, suppose 5-2 is stored in scores.
    # So, scores[0] means the first index of scores, which is 5, score[2] is the third index of scores, which is 2. We will not consider the hyphen('-'). We will only work with numbers for calculations.
    # Now, the 5 and 2 are still strings, So we need covert it for doing calculations.
    # We convert the strings to integers with int() function as we cannot do calculations using strings. We will store the first index (5) which is own team's score in team_score and third index (2) which is enemy team's score in opponent_score.

    team_score = int(scores[0])
    opponent_score = int(scores[2])

    # Check if the team won, lost, or drew the game and update total points and wins accordingly.
    # Win = 3 points. Draw = 1 point and Loss = 0 point.
    if team_score > opponent_score:

        points += 3  # Team gets 3 points for winning.
        wins += 1    # Increment the number of wins by 1.

    elif team_score == opponent_score:

        points += 1  # Team gets 1 point for a draw.

    # We didn't use any extra condition for loss. Because if the team's score is not greater than the opponent's score and also not equal to the opponent's score, then it is assumed to be a loss. Since, no points will be added or deducted for losing, we will not use any else condtion.

# Calculate the winning percentage.
win_percentage = (wins / games) * 100

# Show the results to the user.
print("Total points:", points)
print("Winning percentage:", f"{win_percentage:.2f}%")

# In the line f"{win_percentage:.2f}%", the 'f' before the string means it's an f-string, and 'win_percentage:.2f' means the output will show only two numbers after the decimal point. For example, instead of showing 66.666666666666%, it will show 66.67%.
# In 'win_percentage:.2f', you can determine how many numbers you want to be shown after decimal point by editing the 2 in ':.2f'. If you want to show 3 numbers, you can write :.3f

```
