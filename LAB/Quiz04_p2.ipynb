```python
class Edge:
    def __init__(self, source, destination, next_edge=None):
        self.source = source
        self.destination = destination
        self.next = next_edge
```

```python
class AdjacencyListGraphUndirectedUnweighted:
    def __init__(self, num_vertices):
        self.num_vertices = num_vertices
        self.adj_list = [None] * num_vertices

    def add_edges(self, edges):
        for source, destination in edges:
            self.add_edge(source, destination)
            self.add_edge(destination, source)

    def add_edge(self, source, destination):
        new_edge = Edge(source, destination)
        new_edge.next = self.adj_list[source]
        self.adj_list[source] = new_edge


class AdjacencyListGraphDirectedUnweighted:
    def __init__(self, num_vertices):
        self.num_vertices = num_vertices
        self.adj_list = [None] * num_vertices

    def add_edges(self, edges):
        for source, destination in edges:
            self.add_edge(source, destination)

    def add_edge(self, source, destination):
        new_edge = Edge(source, destination)
        new_edge.next = self.adj_list[source]
        self.adj_list[source] = new_edge
```

```python
def is_team_coordinated(graph):
    """Checks if every directed alert is reciprocated."""

    # Loop through all vertices in the graph
    for u in range(graph.num_vertices):
        current = graph.adj_list[u]  # Get the head of the adjacency list for vertex u

        # Traverse all edges going out from u
        while current != None:
            v = current.destination  # Destination of the current edge u → v
            found = False  # Flag to check if the reverse edge v → u exists

            # Search v's adjacency list for an edge back to u
            temp = graph.adj_list[v]
            while temp != None:
                if temp.destination == u:  # Reverse edge found
                    found = True
                    break
                temp = temp.next  # Move to next edge in v's adjacency list

            if not found:
                # If the reverse edge doesn't exist, coordination is not mutual
                return False

            current = current.next  # Move to the next edge from u

    # If all directed edges have reverse counterparts, return True
    return True



print("==== TEAM COORDINATION CHECK (Directed) ====")

# Graph 1
g1 = AdjacencyListGraphDirectedUnweighted(2)
g1.add_edges([
    (0, 1),  # IronMan → CaptainAmerica
    (1, 0)   # CaptainAmerica → IronMan
])
# 1. IronMan ↔ CaptainAmerica
# Both heroes reciprocated mission alerts.
print("1. Coordination mutual?", is_team_coordinated(g1))  # True

# Graph 2
g2 = AdjacencyListGraphDirectedUnweighted(2)
g2.add_edges([
    (0, 1)   # Thor → Hulk
    # Hulk did not respond
])
# 2. Thor → Hulk (no response)
# Thor contacted Hulk, but Hulk didn’t respond.
print("2. Coordination mutual?", is_team_coordinated(g2))  # False

# Graph 3
g3 = AdjacencyListGraphDirectedUnweighted(3)
g3.add_edges([
    (0, 1),  # BlackWidow → Hawkeye
    (1, 0),  # Hawkeye → BlackWidow
    (1, 2),  # Hawkeye → IronMan
    (2, 1)   # IronMan → Hawkeye
])
# 3. BlackWidow ↔ Hawkeye, Hawkeye ↔ IronMan
# Every communication is mutual.
print("3. Coordination mutual?", is_team_coordinated(g3))  # True
```

```python
def is_collaboration_mutual(graph):
    """Checks if all help connections are mutual between vertices in a directed graph"""

    # Iterate through each vertex u in the graph
    for u in range(graph.num_vertices):
        current = graph.adj_list[u]  # Get the head of u's adjacency list (u's outgoing connections)

        # Traverse each neighbor v that u is helping
        while current is not None:
            v = current.destination  # v is the vertex u is helping
            found = False  # Flag to check if v is also helping u in return

            # Traverse v's adjacency list to look for a connection back to u
            temp = graph.adj_list[v]
            while temp is not None:
                if temp.destination == u:
                    found = True  # Found that v is also helping u
                    break
                temp = temp.next  # Move to the next connection from v

            if not found:
                # If v is not helping u back, then the collaboration is not mutual
                return False

            current = current.next  # Move to u's next help connection

    # If we finish checking all help connections and they all go both ways
    return True

print("\n==== COLLABORATION MUTUAL CHECK ====")

# Graph 1
graph1 = AdjacencyListGraphDirectedUnweighted(3)
graph1.add_edges([(0, 1), (0, 2), (1, 0), (2, 0)])  # Alex→Jordan, Sam; Jordan/Sam→Alex
print("1. Collaboration mutual?", is_collaboration_mutual(graph1))  # True

# Graph 2
graph2 = AdjacencyListGraphDirectedUnweighted(7)
graph2.add_edges([(3, 4), (3, 5), (3, 6), (4, 3), (6, 3)])  # Mia→Liam,Noah,Ella; Liam/Ella→Mia
print("2. Collaboration mutual?", is_collaboration_mutual(graph2))  # False

# Graph 3
graph3 = AdjacencyListGraphDirectedUnweighted(12)
graph3.add_edges([
    (7, 8), (7, 5), (8, 7),         # Ava↔Emma, Ava→Noah
    (5, 7), (5, 9), (9, 5),         # Noah↔Ava, Noah↔Olivia
    (10, 11), (11, 10)              # Lily↔James
])
print("3. Collaboration mutual?", is_collaboration_mutual(graph3))  # True
```

```python
class WEdge:
    def __init__(self, source, destination, weight, next_node=None):
        self.source = source
        self.destination = destination
        self.weight = weight
        self.next = next_node

class AdjacencyListGraph:
    def __init__(self, num_vertices):
        self.num_vertices = num_vertices
        self.adj_list = [None] * num_vertices  # Initialize the adjacency list for each vertex

    def add_edges(self, edges):
        """Add edges with weights to the adjacency list for a directed graph."""
        for source, destination, weight in edges:
            self.add_edge(source, destination, weight) # u to v

    def add_edge(self, source, destination, weight):
        """Helper method to add an edge from source to destination with the given weight."""
        ## Prepend to have insertion in O(1) time

        # Create a new edge
        new_edge = WEdge(source, destination, weight)
        # Insert the edge at the start of the adjacency list for the source vertex
        new_edge.next = self.adj_list[source]
        self.adj_list[source] = new_edge
```

```python
def has_notification_spam(graph):
    """Detect if there is spam (multiple messages sent to the same teammate)."""
    for student in range(graph.num_vertices):
        visited = []  # List to track the destination students that the current student sent messages to
        current_edge = graph.adj_list[student]

        while current_edge:
            destination = current_edge.destination
            weight = current_edge.weight

            # Check if the student has already sent a message to the same destination
            for v in visited:
                if v == destination:
                    return True  # Spam detected, duplicate message
            visited.append(destination)
            current_edge = current_edge.next
    return False  # No spam detected

# Driver Code for Test Cases

# Test Case 1: Sarah pinged Nina twice
graph1 = AdjacencyListGraph(3)
graph1.add_edges([(0, 1, 1), (0, 1, 2), (1, 2, 1)])  # Sarah → [(Nina, 1), (Nina, 2)], Nina → [(Leo, 1)]

# Test Case 2: All messages are unique
graph2 = AdjacencyListGraph(4)
graph2.add_edges([(0, 1, 1), (1, 0, 1), (2, 3, 1), (3, 2, 1)])  # Tom → [(Jerry, 1)], Jerry → [(Tom, 1)], Spike → [(Tyke, 1)], Tyke → [(Spike, 1)]

# Test Case 3: Emma pinged Liam three times
graph3 = AdjacencyListGraph(2)
graph3.add_edges([(0, 1, 1), (0, 1, 2), (0, 1, 3), (1, 0, 1)])  # Emma → [(Liam, 1), (Liam, 2), (Liam, 3)], Liam → [(Emma, 1)]

# Running test cases and displaying results
print("Test Case 1 Result: ", has_notification_spam(graph1))  # Expected: True
print("Test Case 2 Result: ", has_notification_spam(graph2))  # Expected: False
print("Test Case 3 Result: ", has_notification_spam(graph3))  # Expected: True
```

```python
def has_duplicate_alerts(graph):
    """Detect if there is a panic (multiple alerts sent to the same teammate)."""
    for hero in range(graph.num_vertices):
        visited = []  # List to track teammates the current hero has already alerted
        current_edge = graph.adj_list[hero]

        while current_edge:
            destination = current_edge.destination

            # Check for duplicate alert
            for teammate in visited:
                if teammate == destination:
                    return True  # Duplicate alert found
            visited.append(destination)
            current_edge = current_edge.next

    return False  # No duplicates found

# Driver Code for Test Cases

# Test Case 1: IronMan sent two alerts to Thor
graph1 = AdjacencyListGraph(2)
graph1.add_edges([
    (0, 1, 1),  # IronMan → Thor (alert 1)
    (0, 1, 2),  # IronMan → Thor (alert 2)
    (1, 0, 1),  # Thor → IronMan (alert 1)
    (1, 0, 2)   # Thor → IronMan (alert 2)
])

# Test Case 2: All alerts are unique
graph2 = AdjacencyListGraph(3)
graph2.add_edges([
    (0, 1, 1),  # CaptainAmerica → IronMan (alert 1)
    (0, 2, 1),  # CaptainAmerica → Thor (alert 1)
    (1, 0, 1),  # IronMan → CaptainAmerica (alert 1)
    (2, 0, 1)   # Thor → CaptainAmerica (alert 1)
])

# Test Case 3: BlackWidow sent duplicate alerts to Hawkeye
graph3 = AdjacencyListGraph(5)
graph3.add_edges([
    (0, 1, 1),  # BlackWidow → Hawkeye (alert 1)
    (0, 1, 2),  # BlackWidow → Hawkeye (alert 2)
    (1, 0, 1),  # Hawkeye → BlackWidow (alert 1)
    (1, 2, 1),  # Hawkeye → IronMan (alert 1)
    (2, 0, 1)   # IronMan → BlackWidow (alert 1)
])

# Running test cases and displaying results
print("Test Case 1 Result: ", has_notification_spam(graph1))  # Expected: True
print("Test Case 2 Result: ", has_notification_spam(graph2))  # Expected: False
print("Test Case 3 Result: ", has_notification_spam(graph3))  # Expected: True
```
