```python
#Task1

class CellPackage:
  def __init__(self,price,data,talk_time,messages,cashback,validity):
    self.price = int(price)
    self.data = int(data[0:len(data)-3])*1024
    self.talk_time = int(talk_time)
    self.messages = int(messages)
    self.cashback = int(self.price*(int(cashback[0:len(cashback)-1])/100))
    self.validity = validity
  def printDetails(self):
    if self.data != 0:
      print(f'Data = {self.data} MB')
    if self.talk_time != 0:
      print(f'Talktime = {self.talk_time} Minutes')
    if self.messages != 0:
      print(f'SMS/MMS = {self.messages}')
    print(f'Validity = {self.validity} days')
    print(f'--> Price = {self.price} tk')
    if self.cashback != 0:
      print(f'Buy now to get {self.cashback} tk cashback')

# Driver Code

pkg = CellPackage(150, '6 GB', 99, 20, '7%', 7)
print('============= Package 1 =============')
pkg.printDetails()

pkg2 = CellPackage(700, '35 GB', 700, 0, '10%', 30)
print('============= Package 2 =============')
pkg2.printDetails()

pkg4 = CellPackage(120, '0 GB', 190, 0, '0%', 10)
print('============= Package 3 =============')
pkg4.printDetails()
```

```python
#Task2 (Part A)

class box:
  def __init__(self,list1):

    self.height = list1[0]
    self.width = list1[1]
    self.breadth = list1[2]
    print('Creating a Box!')
  def boxDescription(self):
    print(f'Height: {self.height}\nWidth: {self.width}\nBreadth: {self.breadth}')
    self.v = self.height*self.width*self.breadth
  def volume(self):
    return f'Volume of the box is {self.v} cubic units'

# Driver Code
print("Box 1")
b1 = box([10,10,10])
print("=========================")
b1.boxDescription()
print(b1.volume())
print("-------------------------")
print("Box 2")
b2 = box((30,10,10))
print("=========================")
b2.boxDescription()
print(b2.volume())
b2.height = 300
print("Updating Box 2!")
print("Height:", b2.height)
print("Width:", b2.width)
print("Breadth:", b2.breadth)
volume = b2.height * b2.width * b2.breadth
print(f"Volume of the box is {volume} cubic units.")
print("-------------------------")
print("Box 3")
b3 = b2
b3.boxDescription()
print(b3.volume())
```

```python
#Task4

class abcTech:
  def __init__(self,name,designation,department):
    self.name = name
    self.desig = designation
    self.dep = department
    self.ps = []
    self.fw = []
    self.ps_string = ''
    self.fw_string = ''
    print(f'Welcome to abcTech, {self.name}!')

  def addProgrammingSkills(self,languages):
    self.ps += languages

  def addFrameworks(self,frames):
    self.fw += frames

  def printInfo(self):
    print(f'Name: {self.name}')
    print(f'Designation: {self.desig}')
    print(f'Department: {self.dep}')

    for i in range(len(self.ps)):
      if i != len(self.ps)-1:
        self.ps_string += self.ps[i] + ', '
      else:
        self.ps_string += self.ps[i]
    print(f'Programming Skills: {self.ps_string}')

    for j in range(len(self.fw)):
      if j != len(self.fw)-1:
        self.fw_string += self.fw[j] + ', '
      else:
        self.fw_string += self.fw[j]
    print(f'Frameworks: {self.fw_string}')

  def calculateSalary(self,bs,wh):
    self.base_salary = bs
    self.work_hour = wh
    if self.work_hour > 144:
      self.base_salary += (self.work_hour-144)*800
    return self.base_salary

#DriverCode
print("-------------------------")
b1 =abcTech("Tamim Hasan", "Software Engineer", "Android Development")
print("-------------------------")
b1.addProgrammingSkills(["Java", "Python"])
b1.addProgrammingSkills(["Dart", "C++"])
b1.addFrameworks(["Express.js", "React"])
b1.printInfo()
print("-------------------------")
print(f"Your salary for this month is Tk. {b1.calculateSalary(45000, 156)}")
print("-------------------------")
print("-------------------------")
b2 =abcTech("Jahin Khandoker", "Senior Developer", "App Development")
print("-------------------------")
b2.addProgrammingSkills(["Java", "Dart", "Swift"])
b2.addFrameworks(["Flutter", "React Native"])
b2.addFrameworks(["Xamarin"])
b2.printInfo()
print("-------------------------")
print(f"Your salary for this month is Tk. {b2.calculateSalary(103000, 123)}")
print("-------------------------")
```

```python
#Task5

class StudentDatabase:

  def __init__(self,name,ID):
    self.name = name
    self.ID = ID
    self.grades = {}

  def calculateGPA(self,l1,ses):
    l3 = []
    self.GPA = 0
    self.dic1 = {}

    for i in l1:
      l2 = i.split(':')
      l3 += [l2[0]]
      self.GPA += float(l2[1].strip())*3

    self.GPA = (self.GPA/(len(l1)*3))
    self.GPA = round(self.GPA,2)
    l3 = tuple(l3)

    self.dic1[l3] = self.GPA
    self.grades[ses] = self.dic1

  def printDetails(self):
    print(f'Name: {self.name}\nID: {self.ID}')
    for k,v in self.grades.items():
      print(f'Courses taken in {k}')
      for k1,v1 in v.items():
        for i in k1:
          print(i)
        print(f'GPA: {v1}')

#DriverCode
s1 = StudentDatabase('Pietro', '10101222')
s1.calculateGPA(['CSE230: 4.0', 'CSE220: 4.0', 'MAT110: 4.0'], 'Summer2020')
s1.calculateGPA(['CSE250: 3.7', 'CSE330: 4.0'], 'Summer2021')
print(f'Grades for {s1.name}\n{s1.grades}')
print('---------------------------------')
s1.printDetails()
s2 = StudentDatabase('Wanda', '10103332')
s2.calculateGPA(['CSE111: 3.7', 'CSE260: 3.7', 'ENG101: 4.0'], 'Summer2022')
print('---------------------------------')
print(f'Grades for {s2.name}\n{s2.grades}')
print('---------------------------------')
s2.printDetails()
```

```python
#Task6

class Store:
  def __init__(self,balance):
    print('New branch created!')
    self.balance = balance
    self.total_items = 0
    self.dic1 = {}

  def add_item(self,l1):
    self.total_items += 1
    self.dic2 = {}
    self.dic2['stock'] = l1[1]
    self.dic2['buying_price'] = l1[2]
    self.balance -= l1[1]*l1[2]
    self.dic2['selling_price'] = l1[3]
    self.dic1[l1[0]] = self.dic2
    print(f'Item added: {l1[0]}')

  def viewAllItems(self):
    l1 = []
    for i in self.dic1:
      l1 += [i]
    items = ', '.join(l1)               #.join method (Very helpful for turning list into string)
    if l1 == []:
      print(f'There are no items in your inventory')
    else:
      print(f'All items: {items}')

  def viewAllItemDetails(self):
    print(self.dic1)

  def sell_item(self,item1,amount1):
    if self.dic1[item1]['stock'] < amount1:
      print(f'Sorry! {item1} is not available at your desired quantity. Currently we have: {self.dic1[item1]["stock"]}')
    elif self.dic1[item1]['stock'] >= amount1:
      self.dic1[item1]['stock'] = self.dic1[item1]['stock'] - (amount1)
      self.balance += amount1*(self.dic1[item1]['selling_price'])

  def restock_item(self,item2,amount2):
    self.dic1[item2]['stock'] += amount2
    self.balance -= amount2*self.dic1[item2]['buying_price']
    print(f'Restocked item: {item2}, Current Stock: {self.dic1[item2]["stock"]}')


#Drive Code
print("==========================")
branch1 = Store(5000)
print(f"Current Balance: {branch1.balance}")
print(f"Total items: {branch1.total_items}")
branch1.viewAllItems()
branch1.viewAllItemDetails()
print("==========================")
print(f"Current Balance: {branch1.balance}")
branch1.add_item(["ChaCha Noodles", 10, 5, 8])
print(f"Current Balance: {branch1.balance}")
branch1.add_item(["Sparrow Shampoo", 5, 10, 20])
print(f"Current Balance: {branch1.balance}")
print("==========================")
branch1.viewAllItems()
print()
branch1.viewAllItemDetails()
print()
print("==========================")
print(f"Current Balance: {branch1.balance}\n")
branch1.sell_item("ChaCha Noodles", 15)
print(f"Current Balance: {branch1.balance}\n")
branch1.viewAllItemDetails()
print()
branch1.sell_item("ChaCha Noodles", 10)
print()
print(f"Current Balance: {branch1.balance}\n")
branch1.viewAllItemDetails()
print()
print("==========================")
print(f"Current Balance: {branch1.balance}\n")
branch1.restock_item("ChaCha Noodles", 5)
print()
branch1.viewAllItemDetails()
print()
print(f"Current Balance: {branch1.balance}\n")
print("==========================")
```
