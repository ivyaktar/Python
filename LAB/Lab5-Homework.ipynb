```python
#Task2

class Student:
  def __init__(self, name, ID, CGPA):
    self.name = name
    self.__ID = ID
    self.__CGPA = CGPA
  def setId(self,ID):
    self.__ID = ID
  def getID(self):
    return self.__ID
  def getCGPA(self):
    return self.__CGPA


class Department:
  def __init__(self, dep):
    self.dep = dep

    self.s_name = []
    self.s_ID = []
    self.s_CGPA = []

  def findStudent(self,ID):
    for i in range(len(self.s_ID)):
      if self.s_ID[i] == ID:
        print(f'Studen info: \nStudent Name: {self.s_name[i]}\nID: {self.s_ID[i]}\nCGPA: {self.s_CGPA[i]}')
    else:
      print("Student with this ID doesn't exist, Please give a valid ID")

  def addStudent(self, *args):
    for i in args:
      if i.getID() not in self.s_ID:
        self.s_name += [i.name]
        print(f'Welcome to {self.dep} department, {i.name}')
        self.s_ID += [i.getID()]
        self.s_CGPA += [i.getCGPA()]
      else:
        print('Student with the same ID already exists, Please try with another ID')

  def details(self):
    print(f'Department Name: {self.dep}\nNumber of students {len(self.s_name)}\nDetails of the students:')

    for i in range(len(self.s_name)):
      print(f'Student name: {self.s_name[i]}, ID: {self.s_ID[i]}, cgpa: {self.s_CGPA[i]}')

#DriverCode
s1 = Student("Akib", 22301010, 3.29)
s2 = Student("Reza", 22101010, 3.45)
s3 = Student("Ruhan", 23101934, 4.00)
print("1==================================")
cse = Department("CSE")
cse.findStudent(22112233)
print("2==================================")
cse.addStudent(s1,s2,s3)
print("3==================================")
cse.details()
print("4==================================")
cse.findStudent(22301010)
print("5==================================")
s4 = Student("Nakib",22301010,3.22)
cse.addStudent(s4)
print("6==================================")
s4.setId(21201220)
cse.addStudent(s4)
print("7==================================")
cse.details()
print("8==================================")
s5 = Student("Sakib",22201010,2.29)
cse.addStudent(s5)
print("9==================================")
cse.details()
```

```python
#Task3

class Spaceship:
  def __init__(self, name, capacity):
    self.name = name
    self.max_cap = capacity

    self.cargo_list = []
    self.cargo_weight = 0

  def load_cargo(self, cargo):
    self.cargo_list += [cargo.getname()]
    self.cargo_weight += cargo.getweight()

  def display_details(self):
    print(f'Spaceship Name: {self.name}\nCapacity: {self.max_cap}\nCurrent Cargo Weight: {self.cargo_weight}\nCargo: {self.cargo_list}')

class Cargo:
  def __init__(self, name, weight):
    self.__name = name
    self.__weight = weight
  def getname(self):
    return self.__name
  def getweight(self):
    return self.__weight

# Creating spaceships
falcon = Spaceship("Falcon", 50000)
apollo = Spaceship("Apollo", 100000)
enterprise = Spaceship("Enterprise", 220000)
print("1.===================================")
# Creating cargo
gold = Cargo("Gold", 20000)
platinum = Cargo("Platinum", 25000)
dilithium = Cargo("Dilithium", 50000)
trilithium = Cargo("Trilithium", 70000)
neutronium = Cargo("Neutronium", 80000)
print("2.===================================")
# Loading cargo onto spaceships
falcon.load_cargo(gold)
falcon.load_cargo(platinum)
falcon.display_details()
print("3.===================================")
apollo.load_cargo(gold)  # Apollo will not reach its total capacity
apollo.display_details()
print("4.===================================")
falcon.load_cargo(neutronium)  # This should exceed Falcon's capacity
print("5.===================================")
enterprise.load_cargo(dilithium)
enterprise.load_cargo(trilithium)
enterprise.load_cargo(neutronium)  # This should not exceed Enterprise's capacity
enterprise.display_details()
```

```python
#Task4

class Student:
  def __init__(self, name, ID, dep):
    self.name = name
    self.ID = ID
    self.dep = dep

    self.email = None
    self.password = None
    self.login_stat = None
    self.adviced_courses = []

    print('Student object is created!')

class Usis:
  def __init__(self):
    print('USIS is ready to use')

  def login(self, name):
    if name.email == None and name.password == None:
      print('Email and password need to be set')
    else:
      name.login_stat = True
      print('login successful!')

  def advising(self, name, c1 = None, c2 = None, c3 = None, *c ):
    if name.login_stat:
      if len(c) > 0:
        print('You need special approval to take more than 3 courses.')
      elif c1 == None and c2 == None and c3 == None:
        print("You haven't selected any courses")
      else:
        name.adviced_courses = [c1] + [c2] + [c3]
        print('Advising successful!')
    else:
      print('Please logint to advice courses!')

  def individualDetails(self, name):
    courses = ', '.join(name.adviced_courses)
    out = f'Name: {name.name}\nID: {name.ID}\nDepartment: {name.dep}\nAdvised courses: {courses}'
    return out


#DriverCode
rakib = Student("Rakib", 12301455, "CSE")
print("1***********************")
usis_obj = Usis()
print("2***********************")
usis_obj.login(rakib)
print("3***********************")
usis_obj.advising(rakib)
print("4***********************")
rakib.email = "rakib@hotmail.com"
rakib.password = "1234"
print("5***********************")
usis_obj.login(rakib)
print("6***********************")
usis_obj.advising(rakib)
print("7***********************")
usis_obj.advising(rakib, "CSE110", "PHY111", "MAT110", "CSE260")
print("8***********************")
usis_obj.advising(rakib, "CSE110", "PHY111", "MAT110")
print("9***********************")
print(usis_obj.individualDetails(rakib))
```

```python
#Task5

class Train:
  def __init__(self, train, *stops):
    self.train = train
    self.stops = list(stops)
    self.dic = {}
    print(f'Welcome aboard on {self.train}\nStart: {self.stops[0]}\nDestination: {self.stops[len(self.stops)-1]}')

  def addPassenger(self, *pas):
    for i in pas:
      if i.d_start == None:
        i.d_start = self.stops[0]
      if i.d_end == None:
        i.d_end = self.stops[len(self.stops)-1]
      self.dic[i.name] = [i.d_start]+[i.d_end]
      print(f'{i.name} welcome aboard')

  def allPassengerDetails(self):
    for k,v in self.dic.items():
      fair = (self.stops.index(v[1])) - (self.stops.index(v[0]))
      print(f'Name: {k},Start: {v[0]},Destination: {v[1]},Fair: ${fair*100}')

class Passenger:
  def __init__(self, name, d_start = None , d_end = None):
    self.name = name
    self.d_start = d_start
    self.d_end = d_end


#DriverCode
t1 = Train('T1-Express','New York','Manhattan','Brooklyn','Boston')
print("1========================")
p1 = Passenger("Naruto")
t1.addPassenger(p1)
p2 = Passenger("Sasuke","Manhattan")
p3 = Passenger("Hinata","Manhattan","Brooklyn")
print("2========================")
t1.addPassenger(p2,p3)
print("3========================")
t1.allPassengerDetails()
print("4========================")
t2 = Train('Europe-Express','London','Paris','Brussels','Turkey')
print("5========================")
p4 = Passenger("Max","London","Brussels")
p5 = Passenger("Eleven","Paris")
p6 = Passenger("Mike","Brussels")
t2.addPassenger(p4,p5,p6)
print("6========================")
t2.allPassengerDetails()
```

```python
#Task6
class BracuStudent:
  def __init__(self, name, home):
    self.name = name
    self.home = home
    self.bp = 'False'
  def get_pass(self):
    self.bp = 'True'
  def show_details(self):
    print(f'Student Name: {self.name}\nLives in {self.home}\nHave Bus Pass? {self.bp}')

class BracuBus:
  def __init__(self, br, cap = 2):
    self.br = br
    self.cap = cap

    self.pasCount = 0
    self.pas = []

  def board(self, *pas):
    if len(pas) == 0:
      print('No passengers!')
    else:
      for i in pas:
        if i.bp == 'False':
          print("You don't have a bus pass!")
        elif i.home != self.br:
          print('You got on the wrong bus!')
        else:
          if self.pasCount < self.cap:
            print(f'{i.name} boarded the bus.')
            self.pasCount += 1
            self.pas += [i.name]
          else:
            print('Bus is full!')

  def show_details(self):
    print(f'Bus Route: {self.br}\nPassengers Count: {self.pasCount} (Max: {self.cap})\nPassengers On Board: {self.pas}')

#DriverCode
st1 = BracuStudent("Afif", "Mirpur")
print("1===========================")
st2 = BracuStudent("Shanto", "Motijheel")
st3 = BracuStudent("Taskin", "Mirpur")
st1.show_details()
st2.show_details()
print("2===========================")
st3.show_details()
print("3===========================")
bus1 = BracuBus("Mirpur")
bus2 = BracuBus("Azimpur", 5)
bus1.show_details()
bus2.show_details()
print("4===========================")
st2.get_pass()
st3.get_pass()
print("5===========================")
st2.show_details()
st3.show_details()
print("6===========================")
bus1.board()
print("7===========================")
bus1.board(st1, st2)
print("8===========================")
st1.get_pass()
st2.home = "Mirpur"
st1.show_details()
st2.show_details()
print("9===========================")
bus1.board(st1, st2, st3)
print("10===========================")
bus1.show_details()
```
