```python
#Converting all the lowercase elements in the nested list to uppercase
list1 = [['a', 'b', 'c'], ['e', 'f', 'g'], ['x', 'y', 'z']]
for inner_list in list1:
    for index in range(len(inner_list)):
        inner_list[index] = chr(ord(inner_list[index])-32)

print(list1)
```

```python
list1 = ['a', 'b', 'c', 'a', 'a']
for index in range(len(list1)):
    print('range',index)
    print('length',len(list1))
    if list1[index] == 'a':
        index = 0
        print('if',index)
        list1.remove(list1[index])
print(list1)
```

```python
#Correct Solution (Trace and find out why)
list1 = ['a', 'b', 'c', 'a', 'a']
index = 0

while index < len(list1):
  if list1[index] == 'a':
    list1.remove('a')
  else:
    index+=1
print(list1)
```

```python
#Correct Solution 2
list1 = ['a', 'b', 'c', 'a', 'a']

while 'a' in list1:
  list1.remove('a')

print(list1)
```
